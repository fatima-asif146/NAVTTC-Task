### Task:8
Area of Triangle

```python
a=5
b=6
c=7
# calulate semi perimeter
s= (a+b+c)/2
#calclate the area 
area= (s*(s-a)*(s-b)*(s-c))**0.5
print("The area of triangle is:", area)
```
